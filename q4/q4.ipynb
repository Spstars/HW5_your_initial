```python
import csv

def main():
#need to be changed
    file="C:\\Users\\gijin\\HW5_your_initial\\q4\\q4.csv"
    f=open(file,'r',encoding='cp949')
    data=csv.reader(f,delimiter=",")
    head=next(data)
    station1={}
    station2={}
    station3={}
    station4={}
    for d in data:
        if d[1]=="1호선" :
            upPeople=d[4].replace(",","")
            downPeople=d[5].replace(",","")
            customer=int(upPeople)+int(downPeople)
            if d[1] not in station1:
                station1[d[3]]=customer
            else:
                station1[d[3]]+=customer
        elif d[1]=="2호선" :
            upPeople=d[4].replace(",","")
            downPeople=d[5].replace(",","")
            customer=int(upPeople)+int(downPeople)
            if d[1] not in station2:
                station2[d[3]]=customer
            else:
                station2[d[3]]+=customer
        elif d[1]=="3호선" :
            upPeople=d[4].replace(",","")
            downPeople=d[5].replace(",","")
            customer=int(upPeople)+int(downPeople)
            if d[1] not in station3:
                station3[d[3]]=customer
            else:
                station3[d[3]]+=customer
        elif d[1]=="4호선" :
            upPeople=d[4].replace(",","")
            downPeople=d[5].replace(",","")
            customer=int(upPeople)+int(downPeople)
            if d[1] not in station4:
                station4[d[3]]=customer
            else:
                station4[d[3]]+=customer

    station1=sorted(list(station1.items()),key=lambda x:x[1])
    station2=sorted(list(station2.items()),key=lambda x:x[1])
    station3=sorted(list(station3.items()),key=lambda x:x[1])
    station4=sorted(list(station4.items()),key=lambda x:x[1])
    
    print("***Subway Report for Seoul on March 2023***")
    print()
    print("line1")
    print("1st busiest station  : {}".format(station1[-1][0]))
    print("1st least used line : {}".format(station1[0][0]))

    print("line2")
    print("1st busiest station  : {}".format(station2[-1][0]))
    print("1st least used line : {}".format(station2[0][0]))

    print("line3")
    print("1st busiest station  : {}".format(station3[-1][0]))
    print("1st least used line : {}".format(station3[0][0]))

    print("line4")
    print("1st busiest station  : {}".format(station4[-1][0]))
    print("1st least used line : {}".format(station4[0][0]))


    

if __name__ == "__main__":
    main()
```
